# Introduction of Dataset

**About Dataset**
Context
This dataset offers 6,000 Foodpanda records, integrating customer information, order history, payment, and ratings. It displays user behavior towards the platform, restaurants, and meals as well as captures engagement and loyalty patterns. The information is beneficial in analyzing customer churn, spending habits, restaurant and food popularity, and delivery efficiency, along with constructing predictive and recommendation models in the food ordering space.

**Content**
It contains customer demographics like gender, age range, city, signup date, and churn status. It also logs order-level data like restaurant names, dishes, categories, quantities, prices, and delivery results. There are engagement details like order frequency, last order date, and loyalty points, as well as feedback in the form of customer ratings and payment options. Together, it provides a formalized glimpse into customer activity and business performance in a food delivery service.

***About Columns***
customer_id: Unique identifier for each customer.

gender: Gender of the customer (Male, Female, Other).

age: Age group of the customer (e.g., Adult, Senior).

city: City where the customer is located.

signup_date Date when the customer registered on Foodpanda.

order_id Unique identifier for each order.

order_date Date when the order was placed.

restaurant_name Name of the restaurant from which the order was made.

dish_name Name of the dish ordered.

category Category of the food item (e.g., Italian, Dessert).

quantity Number of items ordered.

price Price of the order.

payment_method Payment option used (Cash, Card, Wallet).

order_frequency Total number of orders placed by the customer.

last_order_date Most recent date of order for the customer.

loyalty_points Reward points earned by the customer.

churned Customer activity status (Active or Inactive).

**rating ** Rating given by the customer for the order.

rating_date Date when the rating was submitted.

delivery_status Final status of the delivery (Delivered, Delayed, Cancelled).

```python
plt.figure(figsize = (15,15))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(3, 3, i)
    ax =sns.countplot(data=df, x=col)
    ax.bar_label(ax.containers[0])
    plt.suptitle("Countplot of the columns", fontsize=20, fontweight="bold")  
    plt.xlabel(col.upper())
    plt.ylabel('Count');
```

```python
plt.figure(figsize = (18,18))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(3, 3, i)
    ax =sns.countplot(data=df, x=col, hue= df['churned'])
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    
    plt.xlabel(col)
    plt.ylabel('Count')
```

```python
delivery_status_by_churned = df.groupby("delivery_status", as_index=False)["churned"].value_counts()
print(delivery_status_by_churned)
pivot_delivery = delivery_status_by_churned.pivot(index="delivery_status", columns="churned", values="count")

# Plot
pivot_delivery.plot(kind="bar", figsize=(5,3))
plt.title("Churn Count by Delivery Status")
plt.xlabel("Delivery Status")
plt.ylabel("Count")
plt.legend(title="Churned")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
top_city = df.groupby('city').agg({
    'rating': 'mean',
    'quantity': 'sum',
    'loyalty_points' : 'mean'
}).reset_index()
top_cities = top_city.sort_values(by='rating', ascending=False).head(5)
top_cities
```

```python
top_selling_dish_name = df.groupby('dish_name').agg({
    'rating' : 'mean',
    'quantity':'sum',
    'loyalty_points':'mean'
}).reset_index()
top_selling_dish_name = top_selling_dish_name.sort_values(by='rating', ascending = False).head()
top_selling_dish_name
```

# Feature Engineering

```python
# We can see above that the dates are in object form which we have to convert to date
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['signup_year'] = df['signup_date'].dt.year

df['order_date'] = pd.to_datetime(df['order_date'])
df['order_year'] = df['order_date'].dt.year

df['last_order_date'] = pd.to_datetime(df['last_order_date'])
df['last_order_year'] = df['last_order_date'].dt.year
df.head(3)
```

```python
# maping the features categorial to numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encode_cols = ['gender', 'age','city', 'restaurant_name','dish_name','category','payment_method','churned','delivery_status']
for col in encode_cols:
    df1[col] = le.fit_transform(df1[col])
df1.head(3)
```

```python
plt.figure(figsize = (15, 5))
sns.heatmap(df1.corr(), annot = True)
```
